# Split data / Preprocessing

```python
X: pd.DataFrame = df.drop(columns=["Class"])
y: pd.Series = df['Class']
```

```python
X_res, y_res = SMOTE().fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
lr: LogisticRegression = LogisticRegression(max_iter=10000, random_state=0)
```

```python
model: LogisticRegression = lr.fit(X_train, y_train)
y_pred: np.ndarray = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
y_scores: np.ndarray = model.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
```

```python
auc_score: float = auc(recall, precision)
auc_score
```

```python
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'Precision-Recall Curve (AUC = {auc_score:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
accuracy: float = f"Accuracy: {accuracy_score(y_test, y_pred)}"
precision: float = f"Precision: {precision_score(y_test, y_pred)}"
recall: float = f"Recall: {recall_score(y_test, y_pred)}"
f1: float = f"F1 Score: {f1_score(y_test, y_pred)}"
print(accuracy, precision, recall, f1)
```
